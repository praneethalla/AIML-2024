```python
import pandas as pd
import matplotlib.pyplot as plt

#1. Generate a Data set for Health Care using the following:
#people = [’kiran’, arun’, ’vijay’, ’varun’]
#age =[25, 30, 35, 40, 45]
#height =[145, 151, 165, 173]
#weight=[45, 55, 65, 75]
people = ['kiran', 'arun', 'vijay', 'varun']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]


healthcare_data = pd.DataFrame({
    'Person': people,
    'Age': age,
    'Height': height,
    'Weight': weight
})


print(healthcare_data)

#2. Using the generated dataset given above. Generate the following graphs and justify therelationships among the vectors.

#3. Scatter Plot
#Provide Graph Title, labels for X, Y axis with proper justification and explanation of the graph
plt.figure(figsize=(8, 6))
plt.scatter(healthcare_data['Height'], healthcare_data['Weight'], color='blue')
plt.title('Scatter Plot - Height vs Weight')
plt.xlabel('Height (in cm)')
plt.ylabel('Weight (in kg)')
plt.grid(True)
plt.show()

#4. Bar Chart
#Provide Graph Title, labels for X, Y axis with proper justification and explanation of the graph
plt.figure(figsize=(8, 6))
plt.bar(healthcare_data['Person'], healthcare_data['Age'], color='green')
plt.title('Bar Chart - Age distribution')
plt.xlabel('Person')
plt.ylabel('Age')
plt.grid(axis='y')
plt.show()

#5. Histogram
#Provide Graph Title, labels for X, Y axis with proper justification and explanation of the graph
plt.figure(figsize=(8, 6))
plt.hist(healthcare_data['Height'], bins=20, color='orange', edgecolor='black')
plt.title('Histogram - Height Distribution')
plt.xlabel('Height (in cm)')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```
